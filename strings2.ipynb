```python
#Index starts at [0] Norwegian Blue has a range of 13
#[] have 4 uses in Python this is one of them
#If you go out of range, results in error code
#         01234567890123 this is the range for the phrase below including the space
parrot = "Norwegian Blue"

print(parrot)

print(parrot[3])
print(parrot[4])

print() #Provides a space 

print(parrot[3])
print(parrot[6])
print(parrot[8])

#Negative Indexing by using negative values

print()

print(parrot[-11])
print(parrot[-10])
print()
print(parrot[-11])
print(parrot[-8])
print(parrot[-6])

print()


#Slicing (Start, Stop, Step Values)
print(parrot[0:6]) # Norweg (Start with postion 0 and go up to but do not include 6)

print(parrot[3:5])
print(parrot[0:9])
print(parrot[:9]) #Same code as above without including a start value explicitly
#Start defaults to the beginning of a string when left off,
print()
print(parrot[10:14])
print(parrot[10:]) #Starting value, ending value left off, defaults to end of a string

#2/4 use of [] is slicing, 1/4 was indexing

print(parrot[:]) #Prints out the complete string, this strategy may be useful in lists, but not here 

print(parrot[0:6:2]) #Nre
print(parrot[0:6:3]) #Nw

#Above Exmaple: Slice starts at index position 0
#Exends up to but not including position 6
#Step through sequence inn steps of 2


#This is advance code at this point, but it illustrates uses for slicing
number = "9,223,372,036,854,775,807"
print(number[1::4])
separators = number[1::4]
print(separators)
values = "".join(char if char not in separators else " " for char in number).split()
print([int(val) for val in values])
```
